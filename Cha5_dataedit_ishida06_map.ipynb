```python
data = pd.read_csv('realestateinfo_test1.csv')
data.head(3)
```

```python
data0 = data.head(30)
```

```python
makeUrl  = "https://msearch.gsi.go.jp/address-search/AddressSearch?q="
s_quote  = urllib.parse.quote(data0["アドレス"][0])
response = requests.get(makeUrl + s_quote)
map_info = response.json()[0]["geometry"]["coordinates"]
map_info[0]
```

```python
def Map_info(x):
    makeUrl = "https://msearch.gsi.go.jp/address-search/AddressSearch?q="
    s_quote = urllib.parse.quote(x)
    response = requests.get(makeUrl + s_quote)
    map_info = response.json()[0]["geometry"]["coordinates"]
    return map_info[0]

def Map_info2(x):
    makeUrl = "https://msearch.gsi.go.jp/address-search/AddressSearch?q="
    s_quote = urllib.parse.quote(x)
    response = requests.get(makeUrl + s_quote)
    map_info = response.json()[0]["geometry"]["coordinates"]
    return map_info[1]



data0['緯度'] = data0['アドレス'].apply(Map_info)
data0['経度'] = data0['アドレス'].apply(Map_info2)
```
